# Scoren bij sjeik Pan al Andas


hieronder zie je de functies die ik heb gebruik.

```python
import pandas as pd
df = pd.read_csv('fifa_ratings.csv')
df_club_value = df.groupby('club')['value_eur'].sum().reset_index()
print(df_club_value.sort_values(by='value_eur', ascending=False).head(10))
```

```python
df_plot = df_club_value.sort_values(by='value_eur', ascending=False).head(10)
df_plot.plot(x='club', y='value_eur', kind='bar')
```
